# 7.Determine frequent versus non-frequent customers

```python
#Creating median of days since prior order column 
df_ords_prods_merge_derived['median_order_freq'] =df_ords_prods_merge_derived.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
#Non-frequent custome
df_ords_prods_merge_derived.loc[df_ords_prods_merge_derived['median_order_freq'] > 20, 'order_freq_flag'] = 'Non-frequent customer'
```

```python
#Regular customers 
df_ords_prods_merge_derived.loc[(df_ords_prods_merge_derived['median_order_freq'] <= 20) & (df_ords_prods_merge_derived['median_order_freq'] > 10), 'order_freq_flag'] = 'Regular customer'
```

```python
#Frequent customers
df_ords_prods_merge_derived.loc[df_ords_prods_merge_derived['median_order_freq'] <= 10, 'order_freq_flag'] = 'Frequent customer'
```

```python
#Printing frequency
df_ords_prods_merge_derived['order_freq_flag'].value_counts(dropna = False)
```
