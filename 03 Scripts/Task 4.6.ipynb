# Merging two datasets

```python
#Drop  columns in df_orders_combined
df_orders_combined = df_orders_combined.drop(columns = ['_merge'])
```

```python
#Drop unnamed column in df_prods_clean
df_prods_clean= df_prods_clean.drop(columns = ['Unnamed: 0'])
```

```python
#Merge df_prods_clean and df_orders_combined
df_merged = df_prods_clean.merge(df_orders_combined, on = 'product_id', indicator = True)
```
